```python
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torch.utils.data
from torch.utils.data import Dataset

from model import SNN
from data import HippoDataset


available_device = torch.device("cpu")  # torch.device("mps")

time_steps = 8
learning_rate = 0.001
dataset_path = "./data/data3"
batch_size = 32
epochs = 100


train_set = HippoDataset(dataset_path)
train_loader = torch.utils.data.DataLoader(
    train_set,
    batch_size=batch_size,
    # num_workers=1, 
    shuffle=True)


ce_criterion = nn.CrossEntropyLoss()

net  = SNN(time_steps)
net.to(available_device)
optimizer = optim.Adam(net.parameters(), lr=learning_rate)
# scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min')
pbar = tqdm(range(epochs), desc="epoch")
for epoch in pbar:
    # print(f"epoch: {epoch}")
    # Randomly shuffle train
    epoch_loss = 0
    # if (epoch + 1) % 400 == 0:
    #     optimizer.param_groups[0]['lr'] = learning_rate / 10

    # for x, y in train_loader:
    for i, (x, y) in enumerate(train_loader):
        # print(f"step: {i}")
        # zero the parameter gradients
        print(f"x: {x.shape}, y: {y.shape}")

        optimizer.zero_grad()

        # forward + backward + optimize
        logits = net(x)
        loss = ce_criterion(logits, y)
        
        pbar.set_postfix({'loss': loss.item()})
        loss.backward()
        optimizer.step()
        
        epoch_loss += loss.item()
        
```
